## Silver Layer: Cleansed and Transformed Data

### Loading data from delta

```python
payments = spark.read.format("delta").load("/delta/project3/raw/payments")
riders = spark.read.format("delta").load("/delta/project3/raw/riders")
stations = spark.read.format("delta").load("/delta/project3/raw/stations")
trips = spark.read.format("delta").load("/delta/project3/raw/trips")
display(payments)
```

### Create Silve table based on data in the Bronze layer

```python
spark.sql("DROP TABLE IF EXISTS payments")
spark.sql("DROP TABLE IF EXISTS riders")
spark.sql("DROP TABLE IF EXISTS stations")
spark.sql("DROP TABLE IF EXISTS trips")
```

```python
# Save as Table
spark.sql("CREATE TABLE payments USING DELTA LOCATION '/delta/project3/raw/payments'")
spark.sql("CREATE TABLE riders USING DELTA LOCATION '/delta/project3/raw/riders'")
spark.sql("CREATE TABLE stations USING DELTA LOCATION '/delta/project3/raw/stations'")
spark.sql("CREATE TABLE trips USING DELTA LOCATION '/delta/project3/raw/trips'")
```
